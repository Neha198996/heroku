```python
import tensorflow
import tensorflow_hub as hub
import tensorflow_text as text
hub_preprocess_url='https://kaggle.com/models/tensorflow/bert/frameworks/TensorFlow2/variations/en-cased-preprocess/versions/3'
hub_encoder_url='https://www.kaggle.com/models/tensorflow/bert/frameworks/TensorFlow2/variations/en-cased-l-12-h-768-a-12/versions/4'
bert_preprocesses=hub.KerasLayer(hub_preprocess_url)
bert_encoder=hub.KerasLayer(hub_encoder_url)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import nltk
from nltk.corpus import stopwords
import string
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
import sklearn
```

```python
sw=stopwords.words('english')
sw
```

```python
lemma=WordNetLemmatizer()
```

```python
df=pd.read_csv('/content/airline.csv',encoding='latin')
```

```python
df_New=df[['airline_sentiment','text']]
df_New
```

```python
df_New['text']=df_New['text'].apply(lambda x:x.lower())
```

```python
def Y_test(x):
    if x=='negative' :
        return 0
    elif x=='neutral':
        return 1
    else:
        return 2
```

```python
df_New_X=df_New[['text']]
df_New_Y=df_New['airline_sentiment'].apply(lambda x: Y_test(x))
df_New_Y=pd.DataFrame(df_New_Y)
df_New_Y
```

```python
df_New_X['text']=df_New_X['text'].apply(lambda x:' '.join([word for word in x.split() if word not in sw]))
df_New_X['text']=df_New_X['text'].apply(lambda x:' '.join(word for word in nltk.word_tokenize(x) if word not in string.punctuation))
df_New_X['text']=df_New_X['text'].apply(lambda x:' '.join(word for word in nltk.word_tokenize(x) if word not in 'virginamerica'))
df_New_X['text']=df_New_X['text'].apply(lambda x:' '.join(lemma.lemmatize(word,pos='v') for word in nltk.word_tokenize(x)))
```

```python
X_train,X_test,y_train,y_test=train_test_split(df_New_X,df_New_Y,test_size=0.2,random_state=101)
```

```python
X_train
```

```python
import tensorflow as tf
preprocessed_tex=bert_preprocesses(X_train.text)
```

```python
preprocessed_text_test=bert_preprocesses(X_test.text)
```

```python
bert_model=hub.KerasLayer(hub_encoder_url)
```

```python
bert_results=bert_model(preprocessed_tex)
bert_results.keys()
```

```python
bert_results_test=bert_model(preprocessed_text_test)
```

```python
from tensorflow.keras.layers import Dense,Dropout,LSTM
from tensorflow.keras import models, layers, optimizers,losses
from tensorflow.keras.models import Sequential
model=Sequential()
```

```python
#model.add(LSTM(100,dtype='float32'))
model.add(Dense(10,activation='relu'))
#model.add(Dense(10,activation='tanh'))
```

```python
#model.add(Dropout(0.2))
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics='precision')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping(patience=10, monitor='val_loss', mode='min')

model.fit(bert_results['pooled_output'], y_train, epochs=200, batch_size=45)
```
